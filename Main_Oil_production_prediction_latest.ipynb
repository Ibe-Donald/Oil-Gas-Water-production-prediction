Evlauting the model performance

```python
def nse(observed, predicted):
    return 1 - (np.sum((observed - predicted) ** 2) / np.sum((observed - np.mean(observed)) ** 2))
```

```python
def willmott_index(observed, predicted):
    return 1 - (np.sum((observed - predicted) ** 2) / np.sum((np.abs(predicted - np.mean(observed)) + np.abs(observed - np.mean(observed))) ** 2))
```

```python
def aad(observed, predicted):
    return np.mean(np.abs(observed - predicted))
```

For Linear Regression

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
for i, target in enumerate(y.columns):
    mse = mean_squared_error(y_test.iloc[:, i], y_pred_lr[:, i])
    r2 = r2_score(y_test.iloc[:, i], y_pred_lr[:, i])
    mae = mean_absolute_error(y_test.iloc[:, i], y_pred_lr[:, i])
    nse_value = nse(y_test.iloc[:, i], y_pred_lr[:, i])
    wi_value = willmott_index(y_test.iloc[:, i], y_pred_lr[:, i])
    aad_value = aad(y_test.iloc[:, i], y_pred_lr[:, i])
    print(f"{target}:")
    print(f"MSE: {mse}")
    print(f"R2 Score: {r2}")
    print(f"MAE: {mae}")
    print(f"NSE: {nse_value}")
    print(f"Willmott Index: {wi_value}")
    print(f"AAD: {aad_value}")
    print()
```

Random Forest

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2,5, 6, 8, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]

# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
# Create a base model
rf = RandomForestRegressor(random_state= 42)

# Random search of parameters, using 3-fold cross validation,
# search across 30 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 10, cv = 3, verbose=2, n_jobs = -1, random_state= 42)

# Fit the random search model
rf_random.fit(X_train, y_train)
```

```python
print(rf_random.best_params_)
best_model_vrf = rf_random.best_estimator_
```

```python
# Predict on test set
y_pred_vrf = best_model_vrf.predict(X_test)

# Calculate MSE and R2 score for each target variable
for i, target in enumerate(y.columns):
    mse = mean_squared_error(y_test.iloc[:, i], y_pred_vrf[:, i])
    r2 = r2_score(y_test.iloc[:, i], y_pred_vrf[:, i])
    mae = mean_absolute_error(y_test.iloc[:, i], y_pred_vrf[:, i])
    nse_value = nse(y_test.iloc[:, i], y_pred_vrf[:, i])
    wi_value = willmott_index(y_test.iloc[:, i], y_pred_vrf[:, i])
    aad_value = aad(y_test.iloc[:, i], y_pred_vrf[:, i])
    print(f"{target}:")
    print(f"MSE: {mse}")
    print(f"R2 Score: {r2}")
    print(f"MAE: {mae}")
    print(f"NSE: {nse_value}")
    print(f"Willmott Index: {wi_value}")
    print(f"AAD: {aad_value}")
    print()
```
